```python
import pandas as pd
```

```python
row_data = pd.read_excel("/content/Task-1.xlsx")
price_table = pd.read_excel("/content/Price_table.xlsx")
```

```python
row_data
```

**Removing Rows with Only ID**

```python
row_data.dropna(how='all', subset=row_data.columns.difference(['ID']), inplace=True)
```

```python
row_data
```

```python
row_data['Gender_Code'] = row_data['Gender_Code'].map({'Mr.': 'M', 'Master.': 'M','Mrs.': 'F','Miss.':'F'})
```

 Remove the prefix "age-"

```python
row_data['AGE'] = row_data['AGE'].fillna(0)
row_data['AGE'] = row_data['AGE'].astype(str).str.replace('age-', '', regex=False).astype(int)
```

```python
row_data.isnull().sum()
```

```python
row_data['AGE'] = pd.to_numeric(row_data['AGE'], errors='coerce')
generation_ages = row_data[row_data['AGE'] > 0].groupby('GENERATION')['AGE'].mean()

row_data['AGE'] = row_data.apply(
    lambda x: generation_ages[x['GENERATION']] if x['AGE'] == 0 else x['AGE'], axis=1
)

row_data['AGE'] = row_data['AGE'].round(0).astype(int)
```

```python
row_data.isnull().sum()
```

```python
row_data
```

```python
unnecessary_columns = ['ADDRESS', 'LOCALITY', 'DRIVER_LICENSE']
row_data = row_data.drop(columns=unnecessary_columns, errors='ignore')
```

```python
row_data.isnull().sum()
```

```python
row_data.to_csv('cleaned_data.csv', index=False)

# Use this code to download the file in Colab
from google.colab import files
files.download('cleaned_data.csv')
```

```python
price_dict = price_table.set_index('Product_Name')['Price'].to_dict()
```

```python
productName = list(price_dict.keys())
```

```python
row_data['Total Purchase'] = row_data[productName].mul(price_table.set_index('Product_Name')['Price'], axis=1).sum(axis=1)
```

```python
row_data
```

```python
final_columns = [
    'ID', 'CUST_NAME', 'Gender_Code', 'CITY', 'EMAIL_ADDRESS', 'CREDITCARD_TYPE',
    'ORDER_SHIP_DATE', 'AGE', 'T_TYPE', 'PURCHASE_TOUCHPOINT', 'PURCHASE_STATUS',
    'ORDER_TYPE', 'GENERATION', 'Total Purchase'
]
clean_data = row_data[final_columns]
```
